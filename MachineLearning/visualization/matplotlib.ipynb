```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4])
y=np.array([3,8,1,10])
plt.plot(x,y)
```

```python
x1=np.array([3,8,10,5,7])
plt.plot(x1)
```

```python
x=[1,2,3,4]
y=[3,8,1,10]
plt.plot(x,y,'*')#s- squre, o-circle , .
```

```python
x1=np.array([1,2,3,4,5,6])
y1=np.array([3,8,1,10,5,7])
plt.plot(x1,y1,marker='o')
```

```python
plt.plot(x1,y1,'*:r')
#line - solid, dotted : , dashed --
#color - red r, green g, black k, 
```

```python
plt.plot(x1,y1,'o--g')
```

```python
plt.plot(x1,y1,'s-k')
```

```python
plt.plot(x1,y1,marker='o',ms=15)
#ms: marker size
```

```python
plt.plot(x1,y1,marker='o',ms=15,mfc='r')
#ms: marker size, mfc: marker face color
```

```python
plt.plot(x1,y1,marker='o',ms=15,mec='k',mfc='r')
```

```python
plt.plot(x1,y1,marker='o',linestyle='dotted')
```

```python
plt.plot(x1,y1,marker='o',linestyle='dotted',linewidth=4)
```
